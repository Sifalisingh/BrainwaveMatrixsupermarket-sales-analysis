```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV file ko load karo
data = pd.read_csv("Downloads/supermarket_sales.csv")

# Total sales
data['Total'] = data['Unit price'] * data['Quantity']
total_sales = data['Total'].sum()
print("Total Sales:", total_sales)

# Top product lines
top_products = data.groupby('Product line')['Total'].sum().sort_values(ascending=False)
print("\nTop Product Lines:\n", top_products)

# City-wise sales
city_sales = data.groupby('City')['Total'].sum()
print("\nCity-wise Sales:\n", city_sales)

# Bar chart
top_products.plot(kind='bar', title='Sales by Product Line')
plt.ylabel('Total Sales')
plt.show()

# Pie chart
data['Payment'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Payment Method Share')
plt.ylabel('')
plt.show()
```
